```python

# 🔷 Scenario 1: AI Customer Support Multi-Agent System
# 🎯 Objective:
# Build a multi-agent customer support system that can classify, delegate, and resolve user queries.

# 💼 Problem Statement
# A company wants to automate its support system using AI agents.

# The system should:

# Understand customer queries
# Route them to the correct department
# Provide responses
# 👥 Agents to Build
# Classifier Agent
# Identifies query type (Billing / Technical / General)
# Billing Agent
# Handles payment/refund queries
# Technical Support Agent
# Handles app/software issues
# Response Agent
# Combines outputs into a final response
# ⚙️ Tasks to Implement
# Multi-agent architecture (at least 3 agents mandatory)
# Task delegation logic
# Basic agent communication (message passing / function calls)
# 💡 Sample Inputs
# “I was charged twice for my subscription”
# “The app crashes when I open it”
# “What are your pricing plans?”
# 🧪 Expected Output
# Correct classification of query
# Delegation to appropriate agent
# Final structured response
# 🧰 Tools (Choose Any)
# Python (basic functions or LLM APIs)
# LangGraph (workflow)
# CrewAI (role-based agents)
# AutoGen (agent conversation)
# Microsoft Copilot Studio (optional low-code)
# 📦 Deliverables
# Working code / flow
# Architecture diagram
# 2–3 test case outputs
```

```python
import os
import json
from groq import Groq
from dotenv import load_dotenv


# SETUP


load_dotenv()

api_key = os.getenv("GROQ_API_KEY")

if not api_key:
    raise ValueError(" GROQ_API_KEY not found in .env")

client = Groq(api_key=api_key)


# LLM CALL FUNCTION

def call_llm(prompt):
    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",  # fast & free on Groq
        messages=[
            {"role": "system", "content": "You are a helpful AI assistant."},
            {"role": "user", "content": prompt}
        ],
        temperature=0
    )
    return response.choices[0].message.content.strip()



#AGENT 1: CLASSIFIER

def classifier_agent(query):
    prompt = f"""
    Classify the following customer query into one of:
    - Billing
    - Technical
    - General

    Query: "{query}"

    Only return one word.
    """
    return call_llm(prompt)



# AGENT 2: BILLING

def billing_agent(query):
    prompt = f"""
    You are a billing support agent.

    Handle:
    "{query}"

    Give a professional response.
    """
    return call_llm(prompt)



#  AGENT 3: TECH SUPPORT

def technical_agent(query):
    prompt = f"""
    You are a technical support agent.

    Resolve:
    "{query}"

    Give step-by-step troubleshooting.
    """
    return call_llm(prompt)


#  AGENT 4: GENERAL

def general_agent(query):
    prompt = f"""
    You are a general support assistant.

    Answer:
    "{query}"
    """
    return call_llm(prompt)



# RESPONSE AGENT

def response_agent(query, category, agent_response):
    prompt = f"""
    Create a structured customer support response.

    Query: {query}
    Category: {category}
    Agent Response: {agent_response}

    Format:
    Issue Type:
    Resolution:
    Closing Note:
    """
    return call_llm(prompt)


# ORCHESTRATOR

def support_system(query):
    print(f"\n User Query: {query}")

    # Step 1: Classification
    category = classifier_agent(query)
    print(f" Classified as: {category}")

    # Step 2: Delegation
    if "Billing" in category:
        agent_output = billing_agent(query)

    elif "Technical" in category:
        agent_output = technical_agent(query)

    else:
        agent_output = general_agent(query)

    # Step 3: Final Response
    final_response = response_agent(query, category, agent_output)

    return final_response



# TEST CASES

if __name__ == "__main__":
    queries = [
        "I was charged twice for my subscription",
        "The app crashes when I open it",
        "What are your pricing plans?"
    ]

    for q in queries:
        result = support_system(q)
        print("\n Final Response:\n", result)
        print("-" * 60)
```

```python
# WORKFLOW DIAGRAM:-

# User
#  ↓
# Classifier Agent (LLM)
#  ↓

# Billing , Tech , General

#  ↓
# Specialized Agent
#  ↓
# Response Agent (LLM)
#  ↓
# Final Output
```

```python
# Explanation of Multi-Agent Design

# 1.The system uses a multi-agent architecture where each agent performs a specialized task.
# 2.The Classifier Agent (LLM) identifies query type.
# 3.Based on classification, the Orchestrator delegates the query to the appropriate agent.
# 4.Each agent processes the query independently.
# 5.The Response Agent consolidates outputs into a structured format.
# 6.This modular design improves scalability and maintainability.
```

```python
# Task Delegation Logic:-


# 1.If category == Billing → Billing Agent
# 2.If category == Technical → Technical Agent
# 3.Else → General Agent

# This logic is handled by the orchestrator
# Enables dynamic routing of queries
```

```python
# Agent Communication:-

# 1.Agents communicate via function calls
# 2.Output of one agent becomes input to another
# Example flow:
# User Query → Classifier → (category)
# → Passed to specific agent → Response Agent
```
