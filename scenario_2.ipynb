```python
#  Scenario 2: AI Career Advisor & Learning Path Generator
#  Objective:
# Build a multi-step intelligent advisor system using agent workflows.

#  Problem Statement
# Students need guidance on career paths based on their skills.

# Your system should:

# Analyze student profile
# Suggest career roles
# Generate learning roadmap
# Agents to Build
# Profile Analyzer Agent
# Understands student background
# Career Recommendation Agent
# Suggests suitable roles
# Skill Gap Agent
# Identifies missing skills
# Learning Path Agent
# Creates step-by-step roadmap
# Tasks to Implement
# Sequential workflow (LangGraph preferred)
# State passing between agents
# Conditional logic (beginner vs intermediate)
# Sample Input
# “I know Python basics and statistics, and I want to become a Data Scientist”

# Expected Output
# Suggested roles (e.g., Data Scientist, Analyst)
# Skill gaps (ML, SQL, etc.)
# Learning roadmap (step-by-step)
# Tools (Choose Any)
# LangGraph (recommended)
# CrewAI
# AutoGen
# Python logic-based workflow
# Microsoft Copilot Studio
# Deliverables
# Workflow implementation
# Output for at least 2 student profiles
# Diagram of agent flow
```

```python
import os
from groq import Groq
from dotenv import load_dotenv


# Setup

load_dotenv()
client = Groq(api_key=os.getenv("GROQ_API_KEY"))

def call_llm(prompt):
    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[
            {"role": "system", "content": "You are a helpful career advisor AI."},
            {"role": "user", "content": prompt}
        ],
        temperature=0.3
    )
    return response.choices[0].message.content.strip()



# STATE OBJECT

class State:
    def __init__(self, user_input):
        self.user_input = user_input
        self.profile = None
        self.roles = None
        self.skill_gaps = None
        self.learning_path = None



# AGENT 1: PROFILE ANALYZER

def profile_analyzer(state):
    prompt = f"""
    Analyze this student profile and extract:
    - Skills
    - Experience level (Beginner / Intermediate)

    Input: {state.user_input}

    Output format:
    Skills:
    Level:
    """
    state.profile = call_llm(prompt)
    return state



# AGENT 2: CAREER RECOMMENDER

def career_agent(state):
    prompt = f"""
    Based on this profile:
    {state.profile}

    Suggest 2-3 suitable career roles.
    """
    state.roles = call_llm(prompt)
    return state



#  AGENT 3: SKILL GAP

def skill_gap_agent(state):
    prompt = f"""
    Profile:
    {state.profile}

    Target Roles:
    {state.roles}

    Identify missing skills required.
    """
    state.skill_gaps = call_llm(prompt)
    return state



# AGENT 4: LEARNING PATH

def learning_path_agent(state):
    prompt = f"""
    Profile:
    {state.profile}

    Missing Skills:
    {state.skill_gaps}

    Create a step-by-step learning roadmap.

    If beginner → start from basics
    If intermediate → skip basics and go advanced
    """
    state.learning_path = call_llm(prompt)
    return state



# WORKFLOW (Sequential)

def run_workflow(user_input):
    state = State(user_input)

    state = profile_analyzer(state)
    state = career_agent(state)
    state = skill_gap_agent(state)
    state = learning_path_agent(state)

    return state



#  TEST CASES

if __name__ == "__main__":
    inputs = [
        "I know Python basics and statistics, and I want to become a Data Scientist",
        "I am good at HTML, CSS, and JavaScript and want a tech career"
    ]

    for inp in inputs:
        result = run_workflow(inp)

        print("\n")
        print(f" Input: {inp}")
        print("\n Profile:\n", result.profile)
        print("\n Roles:\n", result.roles)
        print("\n Skill Gaps:\n", result.skill_gaps)
        print("\n Learning Path:\n", result.learning_path)
```
